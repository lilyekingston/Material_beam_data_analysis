# Materials for Cantilever Beam
Importing dataset on properties of various materials and analyzing them to predict which material is best to use for a cantilever beam.
The cantilever beam is fixed on one end and has a load of 3.0 kN on the opposite end. Beam is 0.1 m in diameter and 2.0 m long. The goal of the design is to select the lightest weight material that will not fail and has a deflection less than 20.0 mm. 

```python
import pandas as pd 
df_mechanicals2 = pd.read_excel(r'C:\Users\lilye\Downloads\mechanicals2.xlsx')
df_mechanicals2
```

```python
# adding column to dataframe for material index
mi = (df_mechanicals2['Modulus'])** 0.5 / df_mechanicals2['Density']
df_mechanicals2['Material Index'] = mi
# adding column to dataframe for specific strength
ss = (df_mechanicals2['Strength']) ** (2/3) / df_mechanicals2['Density']
df_mechanicals2['Specific Strength'] = ss

# sorting values by material type and resetting index
df_mechanicals2 = df_mechanicals2.sort_values('Type')
df_mechanicals2 = df_mechanicals2.reset_index(drop=True)
df_mechanicals2
```

```python
# Calculating the maximum stress for a material before it fails
import numpy as np 

F = 3000 # N
L = 2 #m
M = F*L 
d = 0.1 #m 
I = np.pi * (d**4) / 64
sigma = ((M*(d/2))/I) / 10**6 # putting sigma in units of MPa
print(f'The maximum stress is: {sigma} ')
df_newmechanicals2 = df_mechanicals2.loc[df_mechanicals2['Strength'] < sigma]
df_newmechanicals2 = df_newmechanicals2.reset_index(drop=True)
df_newmechanicals2
```

```python
# creating horizontal bar charts using seaborn
import seaborn as sns

sns.catplot(data=df_newmechanicals2, y='Material',x='Material Index', kind='bar')

sns.catplot(data=df_newmechanicals2, y='Material',x='Specific Strength', kind='bar')
```

```python
# generating a pair plot and heat map for materials that meet the criteria
sns.pairplot(data=df_newmechanicals2)
```

```python
sns.heatmap(df_newmechanicals2.corr(),annot=True)
```

```python
# calculating the values of deflection by adding it to a dataframe 
mod_pa = df_mechanicals2['Modulus'] * (10**9)
deflect = ((F)*(L**3))/((3)*(I)*(mod_pa))

df_mechanicals2_delta = df_mechanicals2
df_mechanicals2_delta['Deflection'] = deflect
df_mechanicals2_delta = df_mechanicals2_delta.loc[df_mechanicals2_delta['Deflection'] < 0.02]
df_mechanicals2_delta = df_mechanicals2_delta.reset_index(drop=True)
df_mechanicals2_delta
```

```python
sns.catplot(data=df_mechanicals2_delta, y='Material',x='Deflection', kind='bar')
```

```python
sns.heatmap(df_mechanicals2_delta.corr(),annot=True)
```

```python
# selecting material with highest material index that still meets the strength requirements.
mi_col = df_newmechanicals2['Material Index']
maxmi = mi_col.max()
material_select = df_newmechanicals2.loc[df_newmechanicals2['Material Index'] == maxmi]
material_select
```
